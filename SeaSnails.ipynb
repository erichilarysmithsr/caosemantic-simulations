# Sea Snails Data Interpretation

## Using matplotlib for Sea Snails Data Interpretation

First of all, matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```
# Use the matplotlib for this code

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([106, 106, 109, 110], [84, 83, 85, 88])
plt.xlabel('Snail Length')
plt.ylabel('Snail Diameter')
plt.title('Sea Snails Data Plot')
plt.show()
```
